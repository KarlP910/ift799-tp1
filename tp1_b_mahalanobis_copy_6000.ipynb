```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 20)
pd.set_option('display.max_columns', 20)
```

```python
df_data = pd.read_csv('data\data.csv')
df_labels = pd.read_csv('data\labels.csv')

#Merge data and labels
df_merge = pd.merge(df_data, df_labels, left_index=True, right_index=True)
df_merge.rename(columns={"Unnamed: 0_x":'ID'}, inplace=True)
df_merge.drop(['Unnamed: 0_y'], axis=1, inplace=True)
df_merge.head()
```

```python
#Calculate the threshold for the percentage of zeros
threshold = 0.3 

# Calculate the maximum allowed number of zeros
max_zeros = len(df_merge) * threshold

# Filter columns with 80% or more zeros
filtered_columns = df_merge.columns[(df_merge == 0).sum() <= max_zeros]

# Create a new DataFrame with selected columns
filtered_data = df_merge[filtered_columns]

# Display the modified DataFrame
filtered_data.head()
```

```python
#Make a dataframe for each value of PRAD LUAD BRCA KIRC COAD
df_prad = filtered_data[filtered_data['Class'] == 'PRAD']
df_luad = filtered_data[filtered_data['Class'] == 'LUAD']
df_brca= filtered_data[filtered_data['Class'] == 'BRCA']
df_kirc = filtered_data[filtered_data['Class'] == 'KIRC']
df_coad = filtered_data[filtered_data['Class'] == 'COAD']

#Remove Class column and ID column
df_coad= df_coad.iloc[:,1:-1]

df_kirc= df_kirc.iloc[:,1:-1]

df_brca= df_brca.iloc[:,1:-1]

df_luad= df_luad.iloc[:,1:-1]

df_prad = df_prad.iloc[:,1:-1]
```

```python
# Calculate the correlation matrix
correlation_matrix_prad = df_prad.corr().abs()

# Set the threshold for correlation
threshold = 0.80

# Find columns with correlation greater than or equal to the threshold
high_corr_columns = np.where(correlation_matrix_prad >= threshold)

# Create a set to store columns to remove
columns_to_remove = set()

# Iterate through the pairs of highly correlated columns
for col1, col2 in zip(*high_corr_columns):
    if col1 != col2 and col1 not in columns_to_remove:
        columns_to_remove.add(col2)

# Remove the identified columns
df_prad_filtered = df_prad.drop(columns=df_prad.columns[list(columns_to_remove)],axis=1)
df_prad_filtered.head()
```

### Calculate intra distance for each cancer type BRCA, PRAD, LUAD, COAD, KIRC

```python
#Calculate mahalanobis distance for intra class PRAD
#np.random.seed(1)

#df_sampled_prad = df_prad_filtered_common.sample(axis=1, n=6000)

df_prad_filtered_mean = df_prad_filtered_common.mean()

df_prad_filtered_cov = df_prad_filtered_common.cov()

df_prad_filtered_cov_inv = np.linalg.inv(df_prad_filtered_cov)

distances = []

for index in  range(0, df_prad_filtered_common.shape[0]):
    diff = df_prad_filtered_common.iloc[index, :].values - df_prad_filtered_mean.values
    distance = np.dot(diff.T, np.dot(df_prad_filtered_cov_inv, diff))
    mahalanobis_distance_abs = np.abs(distance) 
    distances.append(np.sqrt(mahalanobis_distance_abs))

max_intra_dist_prad = np.max(distances)

max_intra_dist_prad
```
